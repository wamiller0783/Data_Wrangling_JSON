## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
f = 'data/world_bank_projects.json'
# load 'data/world_bank_projects.json' into Pandas datafram
json_df = pd.read_json(f)
```

### 1. Find the 10 countries with most projects

```python
# use .value_counts method to list the 10 countries with the most projects
json_df['countryname'].value_counts()[:10]
```

```python
# import matplotlib.pyplot for visualization
import matplotlib.pyplot as plt
```

```python
# visualize counts as a categorical bar plot
json_df['countryname'].value_counts()[:10].plot(kind='barh')
# give the plot a title
plt.title('# of Projects per Country')
# label the axis that needs one
plt.xlabel('Count')
# correct the fact that this displays in ascending order
ax = plt.gca()
ax.invert_yaxis()
# show the plot
plt.show()
```

### 2. Find the top 10 major project themes (using column 'mjtheme_namecode')
I'm going to do this two ways, one of which is easy, but doesn't quite follow the instructions given. I will then do this the more complex way, following the instruction to the letter.

#### Method 1 (cheating?)
This does not use the "mjtheme_namecode" column of the dataframe, as the instructions seem to specify. Because of this, it's a far more simple way of solving the problem. It is therefore more Pythonic.

```python
#load json from file as string
data = json.load(open(f))
# normalize nested string under 'mjtheme_namecode'
namecode_normed_df = json_normalize(data, 'mjtheme_namecode')
```

```python
# Create a dataframe consisting only of codes and non-empty names for the sake of referece
code_ref_df = namecode_normed_df[namecode_normed_df['name'] != ''].drop_duplicates().reset_index(drop=True) 
```

```python
# For each code in the reference data frame, get the number of occurences of that code in the normalized
# 'mjtheme_namecode' dataframe. Add these values to the code reference dataframe as a column.
for code in code_ref_df['code']:
    code_ref_df.loc[code_ref_df['code'] == code, 'count'] = \
    namecode_normed_df[namecode_normed_df['code'] == code]['code'].count()
```

```python
# Display the first 10 rows of the code reference dataframe, sorted descending and with index dropped for clarity.
code_ref_df.set_index('code').sort_values('count', ascending=False)[:10]
```

#### Method 2 (hopefully not cheating)
For this, I'll actually use the column in the dataframe that I loaded the JSON into originally, even though the values contain dicts and this is super inconvenient.

```python
# Initialize empty dictionary
namecode_dict = {}
# Iterate each entry in the column 'mjtheme_namecode', each of which contains a list of dicts.
for entry in json_df['mjtheme_namecode']:
    # Iterate each dictionary item in the list.
    for item in entry:
        # If the code in the item has not yet been encountered, make it the key for a dictionary in namecode_dict.
        if item['code'] not in namecode_dict:
            namecode_dict[item['code']] = {}
        
        # If the name associated with a count has not been added, and a name is present,
        # add the name to its dictionary key.
        if len(item['name']) > 0:
            namecode_dict[item['code']]['name'] = item['name']
        
        # If the code in the item has been counted once before, add to the count in its dictionary key... 
        if 'count' in namecode_dict[item['code']]:
            namecode_dict[item['code']]['count'] += 1
        # else initialize a count, starting at one.
        else:
            namecode_dict[item['code']]['count'] = 1      
```

```python
# Build a dataframe from the namecode_dict dictionary
namecodes_fromdict_df = pd.DataFrame.from_dict(namecode_dict, orient='index')
# Name the index appropriately.
namecodes_fromdict_df.index.name = 'code'
# Display the first 10 entries of the dataframe, sorted descending by count.
namecodes_fromdict_df.sort_values('count', ascending=False)[:10]
```

```python
# visualize counts as a categorical bar plot
namecodes_fromdict_df.sort_values('count')[:10].plot(kind='barh', x='name', y='count')
# give the plot a title
plt.title('Top Ten Project Themes')
# label axes
plt.ylabel('')
plt.xlabel('Count')
# remove the unnecessary legend
plt.gca().get_legend().remove()
# show the plot
plt.show()
```

### 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Reset namecodes_fromdict_df index to avoid issues associated with index data type.
namecodes_fromdict_df.reset_index(inplace=True)
```

```python
# Iterate each entry in the column 'mjtheme_namecode', each of which contains a list of dicts.
for entry in json_df['mjtheme_namecode']:
    # Iterate over each dictionary item in the list.
    for item in entry:
        # If the name associated with a count is not present
        if len(item['name']) == 0:
            # retrieve that name from the dataframe created in step 2 and re-indexed above.
            name = namecodes_fromdict_df[namecodes_fromdict_df['code'] == item['code']]['name'].iloc[0]
            item['name'] = name
```
